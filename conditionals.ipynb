## Conditionals

We might want to do something only if a certain condition is True. 

For example we would like to print a number only if it is larger than 0.

To do that we can use `if` statement:

````python
number = -3

if number > 0:
    print(number, 'is positive')

number = 5

if number > 0:
    print(number, 'is positive')
    
````

The structure of `if` statement is similar to the one we just have seen for `for` statement:

we open the line with `if`, precise the condition and then close it with colon `:`
    
The following lines belonging to the `if` statement must be indended

what is retun by larger than statement here is a boolean value, either True or False

````python
print(-3 > 0)
print(5 > 0)
````

You can combine `for` loops and `if` statements:

````python
numbers = [-3, 4, 2, -3, 0]

for number in numbers:
    if number > 0:
        print(number, 'is positive', number)
````

You can also specify what happens if the condition is not met. You can do it using `else`

````python
numbers = [-3, 4, 2, -3, 0]

for number in numbers:
    if number > 0:
        print(number, 'is positive')
    else:
        print(number, 'is non-positive')
````

And if you wish to make more conditions you can use elif:

````python
numbers = [-3, 4, 2, -3, 0]

for number in numbers:
    if number > 0:
        print(number, 'is positive')        
    elif number < 0:
        print(number, 'is negative')
    else:
        print(number, 'is zero')
````

To equate the value we are using the double equals `==` sign

````python
snacks = ['fruits', 'cookies']

for snack in snacks:
    if snack == 'fruits':
        print('healthy choice')
````

To negate we can use `!=`

````python
numbers = [-3, 4, 2, -3, 0]

for number in numbers:
    if number != 0:
        print(number, 'is not zero')    
````

## AND, OR and NOT

We can combine more than one statement using `and` and `or`

````python
numbers = [-3, 4, 2, -3, 0]

for number in numbers:
    if number > 0 and number % 2 == 0:
        print(number, 'is even positive number')
````

It is also possible to use `is in` to check if an element is in the collection. For example if you want to print only those strings which are made of some substring you can do that:

````python
my_list = ['super', 'subway', 'unsubtle', 'sub', 'title']

for next_string in my_list:
    if 'sub' in next_string:
        print(next_string)
````

## Working with data

````python
import numpy as np
data = np.loadtxt(fname='paris_weather.csv', delimiter=',')

import matplotlib.pyplot as plt
%matplotlib inline
````

````python
plt.plot(data.T)
````

````python
print(np.mean(data, 1))
````

````python
for idx in range(len(data)):
    if (np.mean(data[idx,:])) < 10:
        print('there is a suspicous mean in year with idx', idx)
````

We can ask Python to do something different depending on the value

````python
for idx in range(len(data)):
    if (np.mean(data[idx, :])) > 15.5:
        plt.plot(data[idx, :], color = 'r', alpha = 0.5)
    elif (np.mean(data[idx, :])) > 15:
        plt.plot(data[idx, :], color = 'y', alpha = 0.5)
    elif (np.mean(data[idx, :])) > 10:
        plt.plot(data[idx, :], color = 'b', alpha = 0.5)
    else:
        plt.plot(data[idx, :], color = 'g', alpha = 0.5)
````
